```python
t,f = 0,0
for k,v in data.items():
    # print(k)
    result = json.loads(v)
    # for k1,v1 in result.items():
    #     print(k1)
    if result.__contains__('result') and result['result']=='method call, not focal method':
        print(k)
        if result.__contains__('correct answer') and result['correct answer']=='true':
            t += 1
        elif result.__contains__('correct answer') and result['correct answer']=='false':
            f += 1
        else:
            print("error", k)

print('t:{} f:{}'.format(t,f))
```

结论: 存在函数调用, 并且不是focal method时. t:0 f:215

问题3: 参数是a.b这样的形式时, correct answer是true还是false?

```python
t,f = 0,0
for k,v in data.items():
    result = json.loads(v)
    fieldAccExists = False
    if result.__contains__("assertion"):
        if result['assertion'].startswith('assertEquals'):
            if result.__contains__('arg1'):
                argjson = json.loads(result['arg1'])
                if argjson.__contains__('fieldAccess') and argjson['fieldAccess']==True:
                    fieldAccExists = True
            if result.__contains__('arg2'):
                argjson = json.loads(result['arg2'])
                if argjson.__contains__('fieldAccess') and argjson['fieldAccess']==True:
                    fieldAccExists = True
                fieldAccExists = True
        else:
            if result.__contains__('arg'):
                argjson = json.loads(result['arg'])
                if argjson.__contains__('fieldAccess') and argjson['fieldAccess']==True:
                    fieldAccExists = True
    if fieldAccExists:
        if result.__contains__('correct answer') and result['correct answer']=='true':
            t += 1
        elif result.__contains__('correct answer') and result['correct answer']=='false':
            f += 1

print('t:{} f:{}'.format(t,f))
```
